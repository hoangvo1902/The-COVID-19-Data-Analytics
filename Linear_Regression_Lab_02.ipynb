# Almost all the patient overcome this Disease

## * Import Necessary Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
```

## * Read Data

```python
df = pd.read_csv("Collection_36_days.csv")
df.head(7)
```

## * Delete not Country row

```python
not_legit = ['World', 'Africa', 'Asia', 'Europe', 'North America', 'South America', 'Oceania']

df = df[~df['Country, Other'].isin(not_legit)]
df
```

## * Select Input and Output set

```python
X = df['Total Cases'].tolist()
y = df['Total Recovered'].tolist()
```

## * Fit Data and Return Result of OLS Method

```python
OLS = sm.OLS(y, X).fit()
OLS.summary()
```

## * Visualize Single Linear Regression on Scatterplot

```python
X = df['Total Cases'].tolist()
plt.scatter(X, y)
max_x = df['Total Cases'].max()
min_x = df['Total Cases'].min()
 

x = np.arange(min_x, max_x, 1)
 
y = 0.9547 * x

plt.plot(y, 'r')
plt.title("Scatter of Total Cases on Total Recovered")
plt.xlabel("Total Cases")
plt.ylabel("Total Recovered")
plt.show()
```

- At first the model return the formula: 
  * Total_Recovered = 0.9547 * Total_cases  
- Which means almost every patients would overcome the disease, there is very small number of cases passed away 
- The model return R-squared with value 0.98 means it is really realiable and high p-value convinced us the hypothesis that almost all patient would overcome
